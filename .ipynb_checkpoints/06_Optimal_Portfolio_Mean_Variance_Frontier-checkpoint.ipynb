```python
# Calculate the mean, variance and std from excel
N = 3 # number of stocks in portfolio
import numpy as np
mean_vector = np.array([5,6,7]) # change these values
std_matrix = np.array([[1,0,0],[0,1,0], [0,0,1]]) # change these values
portfolio_means = []
portfolio_variances = []

for i in range(1000):
    weights = np.random.uniform(0,1,N)
    weights = weights/np.sum(weights)
    mean_value = np.dot(weights, mean_vector)
    var_value = np.dot(np.dot(weights.T, std_matrix), weights) 
    print(i, weights, np.sum(weights), mean_value, var_value)
    portfolio_means.append(mean_value)
    portfolio_variances.append(var_value)
```

```python

import matplotlib.pyplot as plt
plt.scatter(portfolio_variances, portfolio_means)
```
