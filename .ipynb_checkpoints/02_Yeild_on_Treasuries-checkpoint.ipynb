### Yahoo Finance

Treasury Yeilds: 
- 10 year (^TNX)
- 5 year (^FVX)
- 13 week (^IRX)
- 30 year(^TYX)

```python
import numpy as np
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def loadData(names = ["^TNX"]):
    temp = None
    for i in names: 
        t10y = yf.Ticker(i)
        t10y.info # print for info
        df = t10y.history(period="max")
        df = df.reset_index()
        df = df.sort_values(by='Date', ascending = False)
        df['year'] = df['Date'].astype(str).str[:4]
        df[i[1:]] = df['Close']
        df = df[['Date', i[1:]]]
        if temp is None: 
            temp = df.copy()
        else: 
            temp = pd.concat([temp,df])
    return temp
```

```python
df = loadData(["^TNX", "^FVX", "^IRX", "^TYX"])
plt.plot(df.Date, df.TNX, c = 'red', label = '10y')
plt.plot(df.Date, df.IRX, c = 'green', label = '13w')
plt.plot(df.Date, df.FVX, c = 'blue', label = '5y')
plt.plot(df.Date, df.TYX, c = 'black', label = '30y')
plt.legend()
plt.show()
```
